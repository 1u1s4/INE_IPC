```python
import re

def genero(palabra: str) -> str:
    VOCALES = 'aeiuoáéíóú'
    palabra = palabra.split()[0]
    ultima_letra = []
    ultima_letra.append(palabra[-1])
    if ultima_letra[0] not in VOCALES: 
        for i in range(1, len(palabra)):
            letra = palabra[-i]
            if letra in VOCALES:
                ultima_letra.append(letra)
                break
    if palabra in ('intervención', 'proteción', 'reparación', 'leche', 'carne'):
        return 'la'
    elif palabra in ('dulces', 'puertas', 'revistas', 'vendas', 'flores', 'clases', 'presentaciones'):
        return 'los'
    elif palabra in ('azúcar', 'agua', 'cubrecama', 'gas'):
        return 'el'
    genero = True # True = masculino
    if ultima_letra[0] == 's':
        plural = True
    else:
        plural = False
    if ultima_letra[-1] == 'a':
        if ultima_letra[0] != 'n':
            genero = False
    if ultima_letra[-1] == 'e':
        if len(ultima_letra) == 0:
            genero = False
    
    if genero:
        if plural:
            return "los"
        else:
            return "el"
    else:
        if plural:
            return "las"
        else:
            return "la"

for gb in nombres:
    print(genero(gb), gb)
```
